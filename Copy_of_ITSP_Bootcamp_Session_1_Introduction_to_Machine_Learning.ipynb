#Conditions

As their name indicates, conditionals are a way to execute code depending on whether a condition is True or False. As in other languages, Python supports `if` and `else` but `else if` is contracted into `elif`, as the example below demonstrates.

```python
my_variable = 5
if my_variable < 0:
  print("negative")
elif my_variable == 0:        #Executed when  if condition is evaluated to false
  print("null")
else: # my_variable > 0       #Executed when all if and elif conditions are evaluated to false
  print("positive")
```

#Loops

Loops are a way to execute a block of code multiple times. There are two main types of loops: while loops and for loops.

```python
my_list=[1,2,3,4]
```

```python
#while Loop
i = 0
while i < len(my_list):
  print(my_list[i])
  i += 1 # equivalent to i = i + 1
```

```python
#for loop
for i in range(len(my_list)):
  print(my_list[i])
```

for loop has the syntax of:

for i in range (start, stop, steps):

`start` refers to the initial value of i, `stop` refers to the upper limit which is not included in loop, and `steps` refers to what will be the steps by which i is incremented after each iteration.

By default, start is considered to be from 0 and steps are considered to be increments of 1

This structure is essentially same as the structure used in c++:

for (int i = start; i < stop; i += steps)

If the goal is simply to iterate over a list, then we can do that directly as follows:

```python
for element in my_list:
  print(element)
```

#Step 3: Training the Model
One of the most important practices in machine learning is to split datasets into training and test sets.

As in, a model will **train on the training set** to learn patterns and then those patterns can be **evaluated on the test set**.

Crucially, a model should **never** see testing data during training.

This is equivalent to a student studying course materials during the semester (training set) and then testing their abilities on the following exam (testing set).

Scikit-learn provides the [`sklearn.model_selection.train_test_split`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) method to split datasets in training and test sets.

> **Note:** A common practice to use an 80/20 or 70/30 or 75/25 split for training/testing data. There is also a third set, known as a validation set (e.g. 70/15/15 for training/validation/test) for hyperparameter tuning on but for now we'll focus on training and test sets.

```python
# 1. Define x and y
# Note the double brackets [[]] for x. scikit-learn expects x to be a 2D array.
x = df[['Expense']]
y = df['Sales']

# 2. Split the data
# test_size=0.2 means 20% of the data will be for testing, and 80% for training.
# random_state=42 is like a 'seed' which ensures that everytime the same state is used, the data will be same.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# 3. Print the shapes to see what happened
print(f"x_train shape (rows, columns): {x_train.shape}")
print(f"y_train shape (rows): {y_train.shape}")
print(f"x_test shape (rows, columns): {x_test.shape}")
print(f"y_test shape (rows): {y_test.shape}")
```

#Choosing a Model:

Choosing a model often depends on the type of problem you're working on.

For example, there are different models that Scikit-Learn recommends whether you're working on a classification or regression problem.

You can see a map breaking down the [different kinds of model options and recommendations in the Scikit-Learn documentation](https://scikit-learn.org/stable/machine_learning_map.html).
Scikit-Learn refers to these models as `estimators`.

We'll be using the Linear Regression Model for this example.

#Step 4: Training the Model

Since we've already loaded the dataset, now all that's left is to create an "object" of the model and `fit` the dataset to the model.

Fitting a model to a dataset involves passing it the data and asking it to figure out the patterns.

If there are labels (supervised learning), the model tries to work out the relationship between the data and the labels.

If there are no labels (unsupervised learning), the model tries to find patterns and group similar samples together.

Most Scikit-Learn models have the [`fit(x, y)`](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html) method built-in, where the `x` parameter is the features and the `y` parameter is the labels.

In our case, we start by fitting a model on the training split (`x_train`, `y_train`).

```python
# Create the model "object"
model = LinearRegression()

# Train the model using our training data
# This is where the model "learns" by finding the best-fit line!
model.fit(x_train, y_train)

print("Model trained successfully!")
```

### What did the model "learn"?

The model has found the "best-fit line." A line is defined by its **slope ($m$)** and **intercept ($b$)**.

  * `y = mx + b`
  * `Sales = m * (Expense) + b`

Let's see what the model found for $m$ and $b$.

```python
# The slope (m) is stored in model.coef_
m = model.coef_[0]

# The intercept (b) is stored in model.intercept_
b = model.intercept_

print(f"The best-fit line is: Sales = {m:.2f} * (Expense) + {b:.2f}")
```

#Step 5: Evaluation

`Evaluation` refers to testing the model to see how well it performs after being trained. Metrics like `R-squared`, `Mean squared error`, `Accuracy`, `F1 score` are used to evaluate different aspects of a model and the evaluation criteria and metrics also differ from model to model.

For a Linear Regression model, `R-squared` is generally considered sufficient as it explains how close of a correlation actually exists between the two quantities.

A model is only considered to be complete once it has been evaluated properly.

```python
# The .score() method returns the R-squared value for regression models
score = model.score(x_train, y_train)

print(f"The R-squared score of our model is: {score:.4f}")
print(f"This means our model can explain {score*100:.2f}% of the 'variance' in the sales data which is a really good score")
```

We should always ensure that the model is not tested on the same dataset on which it was trained. In some models, this might lead to score always being perfect as the model "remembers" the data it was trained on.

This kind of false score is called a case of `overfitting`.

```python
score = model.score(x_test, y_test)

print(f"The R-squared score of our model is: {score:.4f}")
print(f"This means our model can explain {score*100:.2f}% of the 'variance' in the sales data which is a really good score")
```

Another way to actually see the accuracy of the model is to plot the `best fit line` given by the model alongside the actual data.

```python
#Use our model to predict salaries for the x_test data
y_pred = model.predict(x_test)

# Plot the results!
plt.figure(figsize=(6, 6))

# Plot the ACTUAL data points (from our test set)
plt.scatter(x_test, y_test, color='blue', label='Actual Sales')

# Plot our MODEL'S PREDICTIONS (the best-fit line)
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Predicted Sales (Our Model)')

plt.title('Model Performance: Actual vs. Predicted', fontsize=20)
plt.xlabel('Expense', fontsize=14)
plt.ylabel('Sales', fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.legend()
plt.show()
```
